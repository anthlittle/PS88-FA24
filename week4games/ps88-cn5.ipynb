```python
# Importing some libraries
import numpy as np
import matplotlib.pyplot as plt
from datascience import Table
%matplotlib inline
```

Recall in the lecture notebook we had some code to plot the expected utility to grabbing and deferring in the persimmon game.

```python
p = np.arange(0,1, step=.01)
eug = p*0 + (1-p)*4
eud = p*1 + (1-p)*3
plt.ticklabel_format(style='plain')
plt.plot(p,eug, label='Grabbing Expected Utility')
plt.plot(p,eud, label='Deferring Expected Utility')
plt.xlabel('Probability of Grabbing')
plt.ylabel('Expected Utility')
plt.legend()
```
